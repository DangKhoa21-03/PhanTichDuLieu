```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
```

```python
df1 = pd.read_csv("students_mental_health_survey.csv")
display(df1)
```

```python
# Đếm số lần xuất hiện của từng giá trị trong một cột cụ thể
value_counts = {col: df1[col].value_counts() for col in df1.columns}
value_counts
```

```python
# Kiểm tra số lượng dòng trùng lặp
duplicate_rows = df1.duplicated()  # Trả về True cho các dòng trùng lặp
duplicate_count = duplicate_rows.sum()  # Tính tổng số dòng trùng lặp
print("Số dòng trùng lặp:", duplicate_count)
```

```python
# Kiểm tra giá trị bị trống
df1.isnull().sum()
```

```python
df1.describe()
```

```python
# Vì cột CGPA có giá trị bị trống nên cần xử lý bằng cách thay giá trị trống bằng giá trị trung bình 
mean_CGPA = df1['CGPA'].mean()
print(mean_CGPA)
```

```python
# Thay giá trị NaN bằng giá trị trung bình mà không dùng inplace
df1['CGPA'] = df1['CGPA'].fillna(mean_CGPA)

# Kiểm tra kết quả
print("5 giá trị đầu tiên của cột CGPA sau khi xử lý:")
print(df1['CGPA'].head(5))
```

```python
# Lọc chỉ các cột có kiểu dữ liệu int và float
Loc_Cot = df1.select_dtypes(include=['int64', 'float64'])
# Tính độ lệch chuẩn cho các cột số
Do_Lech_Chuan = Loc_Cot.std()
print(Do_Lech_Chuan)
```

```python
# Kiểm các giá trị ngoại lai
# Tính Z-score cho mỗi giá trị trong DataFrame
z_scores = zscore(df1.select_dtypes(include=['float64', 'int64']))  # Chỉ áp dụng cho các cột số

# Xác định các giá trị ngoại lai (Z-score lớn hơn 3 hoặc nhỏ hơn -3)
outliers = (z_scores > 3) | (z_scores < -3)

# Lọc các dòng có ít nhất một giá trị ngoại lai
outliers_data = df1[outliers.any(axis=1)]  # Sử dụng `any(axis=1)` để kiểm tra nếu có bất kỳ giá trị ngoại lai nào trong dòng

# In ra các dòng có giá trị ngoại lai
print(outliers_data)

num_outliers = outliers.any(axis=1).sum()

# In ra số dòng có giá trị ngoại lai
print(f'Total number of rows with outliers: {num_outliers}')
df_cleaned = df1[~outliers.any(axis=1)]  # ~ để loại bỏ các dòng có ngoại lai

# In ra DataFrame đã loại bỏ các dòng có ngoại lai
print("DataFrame after removing outliers:")
print(df_cleaned)
```
